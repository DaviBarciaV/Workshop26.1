```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# 1) gerando os valores entre 20 e 35 graus
temperaturas = np.random.randint(20, 35, size=7)

# 2) criando array para os dias da semana
semana = ['domingo', 'segunda-feira', 'terça-feira', 'quarta-feira', 'quinta-feira', 'sexta-feira', 'sábado']

# 3) printando
print(f"dias da semana: {semana}")
print(f"temperaturas da semana (celsius): {temperaturas}")
```

```python
# 4) criando o dataframe com a coluna "dia" e "temperatura"
df = pd.DataFrame({
    "Dia": semana,
    "Temperatura (°C)": temperaturas
})
df
```

```python
# 5) fazendo a média de temperaturas da semana
media_temp = (df["Temperatura (°C)"].mean())
print(f"A média de temperaturas na semana foi: {media_temp:.0f} °C\n")

# 6) mostrando a temperatura máxima e mínima, juntamente com o respectivo dia
print(df.loc[df["Temperatura (°C)"].idxmax()])
print(df.loc[df["Temperatura (°C)"].idxmin()]) # o df.loc busca a linha inteira pelo índice
```

```python
# 7) definindo o gráfico de barras, tamanho e colunas com matplotlib
df.plot(x = "Dia", y = "Temperatura (°C)", color="green", kind = 'bar')
plt.axhline(media_temp, color='red', linestyle='--', label=f'Média: {media_temp:.1f}°C') # criando linha de média
plt.legend(loc = "upper right") # escolhendo a posição da label

# 8) definindo título e rotacionando eixo x para melhor visualização e printando
plt.title("Temperatura por dia da semana")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 7.1) definindo o gráfico de linha, tamanho e colunas
df.plot(x = "Dia", y = "Temperatura (°C)", color="blue", kind = 'line')
plt.axhline(media_temp, color='red', linestyle='--', label=f'Média: {media_temp:.1f}°C') # criando linha de média
plt.legend(loc = "upper right") # escolhendo a posição da label

# 8.1) definindo título e rotacionando eixo x para melhor visualização e printando
plt.title("Temperatura por dia da semana")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 7.2) definindo o gráfico de área, tamanho e colunas
df.plot(x = "Dia", y = "Temperatura (°C)", color="red", kind = 'area', alpha=0.4)  # o alpha define transparência (no caso, 60%)

# 8.2) definindo título e rotacionando eixo x para melhor visualização e printando
plt.title("Temperatura por dia da semana")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 7.3) definindo o gráfico de barras horizontais, tamanho e colunas
df.plot(x = "Dia", y = "Temperatura (°C)", color="green", kind = 'barh')
plt.axvline(media_temp, color='red', linestyle='--', label=f'Média: {media_temp:.1f}°C') # criando linha de média
plt.legend(loc = "upper right") # escolhendo a posição da label

# 8.3) definindo título e printando
plt.title("Temperatura por dia da semana")
plt.tight_layout()
plt.show()
```
